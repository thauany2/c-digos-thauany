### LOOPING FOR


```python
for numero in range(17):
    print(numero)
```

```python
alunos = ['Nicolas', 'Pedro', 'Thauany', 'Yasmin', 'André', 'Manu', 'Andressa']

for bagre in alunos:
    print(bagre)
```

### WHILE

```python
while True:
    pergunta = input("Você vai viajar amanhã?")
    if pergunta == "sim":
        print("Boa viagem")
        break
```
